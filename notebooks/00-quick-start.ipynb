# 量子時空統一理論 - 快速開始

## 簡介
本筆記本展示如何使用QST理論進行基本計算。

```python
# 導入必要庫
import sys
sys.path.append('../src')

import numpy as np
import matplotlib.pyplot as plt
from core.qst_calculator import QSTCalculator
from core.physics_constants import PhysicalConstants, QSTConstants
```

```python
# 1. 顯示物理常數
print("基本物理常數:")
PhysicalConstants.print_constants()
print()
print("QST理論常數:")
QSTConstants.print_qst_constants()
```

```python
# 2. 計算暗能量密度
calc_eff = QSTCalculator('effective')
omega_de = calc_eff.dark_energy_density()
print(f"暗能量密度: Ω_DE = {omega_de:.6f}")
print(f"目標值: 0.690309")
print(f"誤差: {abs(omega_de - 0.690309)/0.690309*100:.6f}%")
```

```python
# 3. 計算火星時間延遲
calc_local = QSTCalculator('local')
tau_mars = calc_local.mars_time_delay()
print(f"火星時間延遲: {tau_mars:.1f} μs/日")
print(f"目標值: 81.6 μs/日")
print(f"誤差: {abs(tau_mars - 81.6)/81.6*100:.2f}%")
```

```python
# 4. 尺度依賴性示例
masses = np.logspace(3, 30, 100)  # 從1kg到1e30kg
beta_eff = [calc_local.beta_effective(m) for m in masses]

plt.figure(figsize=(10, 6))
plt.plot(masses, beta_eff, 'b-', linewidth=2)
plt.axvline(x=calc_local.params['M_th'], color='r', linestyle='--', label=f'M_th = {calc_local.params["M_th"]:.1e} kg')
plt.xscale('log')
plt.xlabel('質量 [kg]', fontsize=12)
plt.ylabel('β_eff', fontsize=12)
plt.title('尺度依賴耦合函數 β_eff(M)', fontsize=14)
plt.grid(True, alpha=0.3)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# 5. 表面密度-加速度關係
sigma = np.logspace(-3, 2, 100)  # 10^-3 到 10^2
a_ratio = [calc_local.effective_a0_ratio(s) for s in sigma]

plt.figure(figsize=(10, 6))
plt.plot(sigma, a_ratio, 'g-', linewidth=2)
plt.axvline(x=0.1, color='r', linestyle='--', label='σ_crit = 0.1')
plt.axvline(x=10.0, color='orange', linestyle='--', label='σ_transition = 10.0')
plt.xscale('log')
plt.yscale('log')
plt.xlabel('表面密度 σ [10^9 M_sun/kpc²]', fontsize=12)
plt.ylabel('a_eff / a₀', fontsize=12)
plt.title('表面密度-加速度尺度關係 (QST v4.2)', fontsize=14)
plt.grid(True, alpha=0.3, which='both')
plt.legend()
plt.tight_layout()
plt.show()
```
